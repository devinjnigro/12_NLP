# News Headlines Sentiment

Use the news api to pull the latest news articles for bitcoin and ethereum and create a DataFrame of sentiment scores for each coin. 

Use descriptive statistics to answer the following questions:
1. Which coin had the highest mean positive score?
2. Which coin had the highest negative score?
3. Which coin had the highest positive score?

```python
# Read your api key environment variable
api_key=os.environ["NEWSAPI_CLIENT_ID"]
```

```python
# Create a newsapi client
newsapi = NewsApiClient(api_key)
```

```python
n = pickle.load(open('newsapi.pickle','rb'))
```

```python
n
```

```python
current_date = date.today()
past_date = date.today() - timedelta(weeks=4)
```

```python
# Fetch the Bitcoin news articles

def get_articles(keyword):
    all_articles = []
    all_dates = []
    date = current_date
    #print(f"Fetching news about '{keyword}'")
    #print("*" * 30)
    while date > past_date:
        #print(f"retrieving news from: {date}")
        articles = n.get_everything(
            q=keyword,
            from_param=str(date),
            to=str(date),
            language="en",
            sort_by="relevancy",
            page=1,
        )
        content = []
        for i in range(0, len(articles["articles"])):
            content.append(articles["articles"][i]["content"])
        all_articles.append(content)
        all_dates.append(date)
        date = date - timedelta(days=1)
    return all_articles, all_dates
```

```python
bitcoin_articles, b_dates = get_articles('bitcoin')
```

```python
# Fetch the Ethereum news articles
ethereum_articles, e_dates = get_articles('ethereum')
```

```python
def article_sentiment_summarizer_avg(articles):
    compound_sentiment = []
    pos_sentiment = []
    neg_sentiment = []
    neu_sentiment = []
    text = []
    for day in articles:
        day_score_compound = []
        day_score_pos = []
        day_score_neg = []
        day_score_neu = []
        day_text = []
        for h in day:
            if h == None:
                continue
            else:
                day_score_compound.append(analyzer.polarity_scores(h)['compound'])
                day_score_pos.append(analyzer.polarity_scores(h)['pos'])
                day_score_neg.append(analyzer.polarity_scores(h)['neg'])
                day_score_neu.append(analyzer.polarity_scores(h)['neu'])
                day_text.append(h)
        compound_sentiment.append(sum(day_score_compound) / len(day_score_compound))
        pos_sentiment.append(sum(day_score_pos) / len(day_score_pos))
        neg_sentiment.append(sum(day_score_neg) / len(day_score_neg))
        neu_sentiment.append(sum(day_score_neu) / len(day_score_neu))
        text.append(day_text)
    return compound_sentiment, pos_sentiment, neg_sentiment, neu_sentiment, text
```

```python
bitcoin_sentiments = pd.DataFrame(
    {
        "Compound": article_sentiment_summarizer_avg(bitcoin_articles)[0],
        "Positive": article_sentiment_summarizer_avg(bitcoin_articles)[1],
        "Negative": article_sentiment_summarizer_avg(bitcoin_articles)[2],
        "Neutral" : article_sentiment_summarizer_avg(bitcoin_articles)[3],
        "Text": article_sentiment_summarizer_avg(bitcoin_articles)[4]
    }
)
```

```python
def list_to_string(lst):
    strng = ''
    strng = strng.join(lst)
    return strng
```

```python
bitcoin_sentiments['Text'] = [list_to_string(x) for x in bitcoin_sentiments['Text']]
```

# Tokenizer

In this section, you will use NLTK and Python to tokenize the text for each coin. Be sure to:
1. Lowercase each word
2. Remove Punctuation
3. Remove Stopwords

```python
from nltk.tokenize import word_tokenize, sent_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer, PorterStemmer
from string import punctuation
import re
```

```python
lemmatizer = WordNetLemmatizer()
```

```python
# Expand the default stopwords list if necessary
```

```python
# Complete the tokenizer function
def tokenizer(text):
    """Tokenizes text."""
    
    # Remove the stop words
    sw = set(stopwords.words('english'))
    
    # Remove the punctuation
    regex = re.compile("[^a-zA-Z ]")
    re_clean = regex.sub('', text)
    
    # Create a list of the words
    words = word_tokenize(re_clean)
    
    # Lemmatize Words into root words
    lem = [lemmatizer.lemmatize(word) for word in words]
    
    # Convert the words to lowercase
    #words.lower()
    tokens = [word.lower() for word in lem if word.lower() not in sw]
    
    return tokens
```

```python
# Create a new tokens column for bitcoin
bitcoin_sentiments['Tokens'] = [tokenizer(i) for i in bitcoin_sentiments['Text']]
bitcoin_sentiments.head()
```

```python
# Create a new tokens column for ethereum
ethereum_sentiments['Tokens'] = [tokenizer(i) for i in ethereum_sentiments['Text']]
ethereum_sentiments.head()
```

# NGrams and Frequency Analysis

In this section you will look at the ngrams and word frequency for each coin. 

1. Use NLTK to produce the n-grams for N = 2. 
2. List the top 10 words for each coin. 

```python
from collections import Counter
from nltk import ngrams
```

```python
# Generate the Bitcoin N-grams where N=2
def bigram_counter(text):
    big_string = ''.join(text)
    tokens = tokenizer(big_string)
    bigrams = ngrams(tokens, n=2)
    return Counter(bigrams).most_common(20)

bigram_counter(bitcoin_sentiments['Text'])
```

```python
# Generate the Ethereum N-grams where N=2
bigram_counter(ethereum_sentiments['Text'])
```

```python
# Use the token_count function to generate the top 10 words from each coin
def token_count(tokens, N=10):
    """Returns the top N tokens from the frequency count"""
    return Counter(tokens).most_common(N)
```

```python
# Get the top 10 words for Bitcoin
bitcoin_token_list = []
for row in bitcoin_sentiments['Tokens']:
    for token in row:
        bitcoin_token_list.append(token)
        
token_count(bitcoin_token_list)
```

```python
# Get the top 10 words for Ethereum
ethereum_token_list = []
for row in ethereum_sentiments['Tokens']:
    for token in row:
        ethereum_token_list.append(token)

token_count(ethereum_token_list)
```

# Named Entity Recognition

In this section, you will build a named entity recognition model for both coins and visualize the tags using SpaCy.

```python
import spacy
from spacy import displacy
```

```python
# Optional - download a language model for SpaCy
#!python -m spacy download en_core_web_sm
```

```python
# Load the spaCy model
nlp = spacy.load('en_core_web_sm')
```

## Bitcoin NER

```python
# Concatenate all of the bitcoin text together
bitcoin_string = ''
for row in bitcoin_sentiments['Text']:
    bitcoin_string = bitcoin_string + row
    
bitcoin_string[:4000]
```

```python
# Run the NER processor on all of the text
bit_doc = nlp(bitcoin_string[:4000])

# Add a title to the document
bit_doc.user_data["title"] = "Bitcoin NER"
```

```python
# Render the visualization\
displacy.render(bit_doc, style='ent', jupyter=True)
```

```python
# List all Entities
bit_entities = [ent.text for ent in bit_doc.ents]
bit_entities[:200]
```
